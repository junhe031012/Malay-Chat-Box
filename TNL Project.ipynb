```python
import pandas as pd
import re

# STEP 1: Load the CSV file with fallback encoding
df = pd.read_csv('Final_QA.csv', encoding='latin1')  # Change encoding if needed

# STEP 2: Check column names
print("Original Columns:", df.columns.tolist())

# STEP 3: Strip spaces from column names and rename columns if necessary
df.columns = df.columns.str.strip()  # Remove leading/trailing spaces from column names

# If column names are different, rename them:
df = df.rename(columns={
    'Question': 'question',  # Modify based on actual column name
    'Answer (copy-paste from context)': 'answer',  # Modify based on actual column name
    'Context (less than or equal 512 words)': 'context'  # Modify based on actual column name
})

# STEP 4: Drop rows with missing question, context, or answer
df = df.dropna(subset=['question', 'context', 'answer'])

# STEP 5: Clean text function
def clean_text(text):
    text = str(text)
    text = re.sub(r'\s+', ' ', text)  # Replace multiple spaces/newlines with single space
    text = re.sub(r'[^\w\s.,!?;:()\u00C0-\u024F\u1E00-\u1EFF\u0600-\u06FF\u0900-\u097F\u0E00-\u0E7F-]', '', text)  # Remove weird characters
    text = text.strip()
    return text

# Apply text cleaning to the 'question', 'context', and 'answer' columns
df['question'] = df['question'].apply(clean_text)
df['context'] = df['context'].apply(clean_text)
df['answer'] = df['answer'].apply(clean_text)

# STEP 6: Filter overly short or long entries
df = df[df['question'].str.len() > 3]  # Ensure questions are not too short
df = df[df['question'].str.len() < 512]  # Ensure questions are not too long
df = df[df['context'].str.len() > 128]  # Ensure context isn't too long (optional)
df = df[df['answer'].str.len() < 256]  # Ensure answer is reasonable in length

# STEP 7: Remove duplicates
df = df.drop_duplicates(subset=['question', 'context', 'answer'])

# STEP 8: Final check
print("Total cleaned rows:", len(df))

# STEP 9: Save the cleaned data
df.to_csv('cleaned_FQA.csv', index=False)
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from googletrans import Translator
from langdetect import detect

# Read the data from CSV
df = pd.read_csv('cleaned_FQA.csv')
df.columns = df.columns.str.strip()

documents = df['question'].tolist()
answers = df['answer'].tolist()

# Create TF-IDF vectorizer and fit the model
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(documents)

# Function to get the top k most similar answers
def answer_question(question, top_k=1):
    question_vec = vectorizer.transform([question])
    similarities = cosine_similarity(question_vec, tfidf_matrix).flatten()
    top_indices = similarities.argsort()[::-1][:top_k]

    results = []
    for idx in top_indices:
        result = {
            "question": documents[idx],
            "similarity_score": round(similarities[idx], 3),
            "answer": answers[idx]
        }
        results.append(result)
    return results

# Language translation function
def translate_answer(answer, target_lang):
    translator = Translator()
    if target_lang == "en":
        translated = translator.translate(answer, src='ms', dest='en')
    elif target_lang == "ms":
        translated = translator.translate(answer, src='en', dest='ms')
    return translated.text

# Main function to process the input question
def process_question(question, lang_choice):
    # Get answer based on the question
    results = answer_question(question, top_k=3)
    
    # Process and display the answer based on the user's language choice
    for i, r in enumerate(results, 1):
        print(f"\nResult #{i}")
        print(f"Similarity Score: {r['similarity_score']}")
        
        if lang_choice == "en":
            translated_answer = translate_answer(r['answer'], lang_choice)
            print(f"Answer in Malay: {translated_answer}")
        
        elif lang_choice == "ms":
            print(f"Answer in English: {r['answer']}")


# Main program
if __name__ == "__main__":
    lang_choice = input("Choose your language (en for English, ms for Malay): ").lower()
    
    if lang_choice == 'en':
        print("Your Question (contoh: Where is Melaka?):")
    else:
        print("Soalan Anda (contoh: Apa yang menarik di Putrajaya?):")

    question = input("> ")
    process_question(question, lang_choice)
```
